## CSV - PANDAS
En este script se busca obtener datos de un CSV previamente cargado mediante un programa de extraccion de datos de facebook. Usando Pandas se busca leer dicho CSV y filtrarlo para solo mostrar los datos de la publicaciones cuyos dias de publicacion fueron pares, elimminando asi los impares.

### Paso 1
Importamos la libreria pandas y declaramos un par de funciones

```python
import pandas as pd
```

#### funcion get_day_number
Debido a que la fecha de las publicaciones esta almacenada como un string con el formato "dia de mes", necesitaremos identificar primero el valor del dia, para ello uttilizamos el metodo split y separamos el string segun los espacios en blaco (" ") que contenga quedando una lista de esta forma ['dia','de','mes'] y asi tomar el valor de la primera ubicacion como nuestro valor de dia para luego parsearlo a entero con el metodo int

```python
def get_day_number(date_):
    value_day = int(date_.split(' ')[0])
    return value_day
```

#### funcion check_par
Esta funcion la utilizaremos para verificar si es un numero par o impar, retornando False si es impar y True si es par, para ello tomaremos el residuo de una division entre 2, si esta divison no es cero entonces es un numero impar y retornamos False

```python
def check_par(number):
    if number%2 ==  0:
        return True
    else:
        return False
```

### Paso 2
En este paso obtendremos los valores del CSV mediante la libreria panda, obteniendo un dataframe que procederemos a mostrar

```python
df = pd.read_csv('files/Nerismar_Febres_2021-09-13_09_21_00.597943.csv')
```

```python
df
```

### Paso 3
Procederemos a filtrar los datos del dataframe, para ello utiliaremos los metodos creados anteriormente. Primero ubicamos la columna en cuestion que en este caso es date_post, le aplicamos el metodo de get_day_number y seguido de esto el de check_par retornando asi solo la informacion de las fechas de dias pares

```python
df_whit_par = df[df['date_post'].apply(get_day_number).apply(check_par)]
```
